```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
os.getcwd()
```

```python
recon = sorted(glob.glob('studies/logs/l-town-chebnet-pipe_length-minmax-self_loop-1.csv'))
```

```python
df_dict = {key:value for (key,value) in zip(recon,[pd.read_csv(file,index_col='Unnamed: 0') for file in recon])}
```

```python
df_dict
```

```python
fig,ax = plt.subplots(nrows=2, ncols=3, figsize=(16,10), dpi=300)

for line,axis in zip(df_dict[recon[0]].columns, ax.ravel()):
    for key,val in df_dict.items():
        axis.plot(val[line], label = key)
    axis.set_title(line)

handles, labels = axis.get_legend_handles_labels()
by_label        = dict(zip(labels, handles))
# fig.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(0.45, -0.1, 0.5, 0.5),prop={'size': 12})
fig.delaxes(ax[1,2])
plt.show()
```
